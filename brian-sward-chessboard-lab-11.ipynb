# This notebook shows if the Blue Queen can see the Red Opponent

## Do all your imports

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import random as rdm

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
%matplotlib inline
import matplotlib.pyplot as plt
```

## Define class and color attributes

```python
blue = (0, 1, 1)
black = (0, 0, 0)
white = (1, 1, 1)
red = (1, .2, 0)
is_under = (0, .1, 1)
red_spot = []
blue_spot = []

class ChessBoard:
    def __init__(self):
        self.grid = np.zeros((8,8,3))
        
    def add_red(self, row, col):
        new_board.grid[row,col] = red
        red_spot.append(row)
        red_spot.append(col)
    
    def add_blue(self, row, col):
        new_board.grid[row,col] = blue
        blue_spot.append(row)
        blue_spot.append(col)
    
    def render(self):
        plt.imshow(new_board.grid)
    
    def is_under_attack(self, red_lady, blue_lady):
        #rows and cols
        if red_lady[0] == blue_lady[0] or red_lady[1] == blue_lady[1]:
            return True
        if abs((red_lady[1]-blue_lady[1])/(red_lady[0]-blue_lady[0])) == 1:
            return True
        return False
    
    def is_under_attack_proof(self):
        rrooww = blue_spot[0]
        ccooll = blue_spot[1]
        bucket = blue_spot
        #color row
        for i in range(8):
            new_board.grid[i,ccooll] = is_under
        #color col
        for i in range(8):
            new_board.grid[rrooww,i] = is_under
        #color diags
        #color ne
        ne_row = rrooww
        ne_col = ccooll
        for i in range(8-ne_row):
            while ne_row >= 0 and ne_col < 8:
                new_board.grid[ne_row,ne_col] = is_under
                ne_row -= 1
                ne_col += 1
        #color se
        se_row = rrooww
        se_col = ccooll
        for i in range(8-se_row):
            while se_row < 8 and se_col < 8:
                new_board.grid[se_row,se_col] = is_under
                se_row += 1
                se_col += 1
        #color sw
        sw_row = rrooww
        sw_col = ccooll
        for i in range(8-sw_col):
            while sw_row < 8 and sw_col >= 0:
                new_board.grid[sw_row,sw_col] = is_under
                sw_row += 1
                sw_col -= 1
        #color nw
        nw_row = rrooww
        nw_col = ccooll
        for i in range(8-nw_col):
            while nw_row >= 0 and nw_col >= 0:
                new_board.grid[nw_row,nw_col] = is_under
                nw_row -= 1
                nw_col -= 1
              

        new_board.add_blue(blue_spot[0],blue_spot[1])
        
        
```

## Render the Gameboard

### this first bit does the checkboard // second bit does inital coloring, adding the red and blue pieces

```python
new_board = ChessBoard()

for i in range(8):
    for j in range (8):
        if i % 2 == 0 and j % 2 == 0:
            new_board.grid[i,j] = white
        if i % 2 == 1 and j % 2 == 1:
            new_board.grid[i,j] = white

new_board.add_blue(rdm.randint(0,7), rdm.randint(0,7))
new_board.is_under_attack_proof()
new_board.add_red(rdm.randint(0,7), rdm.randint(0,7))

```
